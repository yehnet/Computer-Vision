```python
from sklearn.datasets import fetch_openml


X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(10,activation='relu',input_shape=(784,)))
model.add(Dense(10,activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X = X.values
y = y.astype(int).values

train_X, val_X, train_y, val_y = train_test_split(X,y,test_size= 0.3, random_state=123, stratify=y)

# X_train = X.to_numpy()
# y_train = y.to_numpy()
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical



np.random.seed(123)
model.fit(train_X,to_categorical(train_y),validation_split=0.2,shuffle=True,epochs=30)
```

```python
probas = model.predict(val_X)

pred_y = np.argmax(probas, axis=1)
```

```python
from sklearn.metrics import roc_auc_score

ary_pred = np.zeros((pred_y.shape[0], 10))
for i, val in enumerate(pred_y):
    ary_pred[i, val] = 1

ary_true = np.zeros((val_y.shape[0], 10))
for i, val in enumerate(val_y):
    ary_true[i, val] = 1

r_a_s = roc_auc_score(ary_true, ary_pred, multi_class='ovr')
print(r_a_s)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ary_true, ary_pred)
```
